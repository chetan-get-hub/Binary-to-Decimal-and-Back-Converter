```python
from decimal import Decimal, InvalidOperation, ROUND_HALF_UP


CENT = Decimal("0.01")
COIN_VALUES: dict[str, int] = {
    "quarters": 25,
    "dimes": 10,
    "nickels": 5,
    "pennies": 1,
}


def read_non_negative_money(prompt: str) -> Decimal:
    while True:
        user_input = input(prompt).strip()
        try:
            value = Decimal(user_input)
        except InvalidOperation:
            print("Please enter a valid amount (example: 12.50).")
            continue

        if value < 0:
            print("Please enter a non-negative amount.")
            continue

        return value.quantize(CENT, rounding=ROUND_HALF_UP)


def to_cents(amount: Decimal) -> int:
    return int((amount / CENT).to_integral_value(rounding=ROUND_HALF_UP))


def make_change(change_cents: int) -> dict[str, int]:
    remaining = change_cents
    breakdown: dict[str, int] = {}

    for coin_name, coin_value in COIN_VALUES.items():
        count, remaining = divmod(remaining, coin_value)
        breakdown[coin_name] = count

    return breakdown


def main() -> None:
    print("Change Return Program")

    cost = read_non_negative_money("Enter item cost: ")
    amount_given = read_non_negative_money("Enter amount given: ")

    if amount_given < cost:
        shortfall = cost - amount_given
        print(f"Insufficient amount. Additional ₹{shortfall:.2f} is needed.")
        return

    change = amount_given - cost
    change_cents = to_cents(change)
    coin_breakdown = make_change(change_cents)

    print(f"\nTotal change: ₹{change:.2f}")
    for coin_name in ("quarters", "dimes", "nickels", "pennies"):
        print(f"{coin_name.title()}: {coin_breakdown[coin_name]}")


if __name__ == "__main__":
    main()
```
